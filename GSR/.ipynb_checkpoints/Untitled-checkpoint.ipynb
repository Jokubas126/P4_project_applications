```python
import matplotlib.pyplot as plt
import statistics
import numpy as np
from scipy.signal import find_peaks

#read the data from the file
def read():
	newList = []
	with open('datawhilewatching2.txt', 'r') as file:
		newList = file.readlines()
	#to remove the '\n'
	newList = [line.strip() for line in newList]
	file.close()
	return newList

def rawGSRList(list):
	toFind = ';'
	newDataList = []
	for i in range(200, len(list)): #we skip 3~ minutes of data as it is callibrating
		temp = list[i]
		numb = temp.find(toFind)
		newDataList.append(1000 - int(temp[:(numb)]))
	return newDataList

rawdatalist = rawGSRList(read())

#applying a median filter for each i+4, i-4 data point
#and using it to calculate the phasic signal
def filteredList(list):
	filteredList = [0,0,0,0] #fill first 4 entries with 0 as we wont be finding phasic data for first 4 seconds
	for i in range(len(list)):
		kernItems = []
		if i > 3 and i < (len(list) - 3):
			for j in range(i - 4, i + 4):
				kernItems.append(int(list[j]))
			medianValue = np.median(kernItems)
			#phasic data
			filteredList.append(list[i] - medianValue)

	#print(filteredList)
	return filteredList

#ploting the data
fig,a = plt.subplots(2, figsize=(30,6))
a[0].plot(rawdatalist)
a[1].plot(filteredList(rawdatalist))
plt.show()

```

```python
#Finding peaks

fig = plt.figure(figsize=(30,6))
x = np.array(filteredList(rawdatalist))
peaks4, _ = find_peaks(x, height=1) 
fig,a = plt.subplots(2, figsize=(30,6))
a[0].plot(rawdatalist)
a[1].plot(peaks4, x[peaks4], "ob"); plt.plot(x); plt.legend(['Height'])
plt.show()
peaks4
```

```python
#Finding peaks, onsets and offsets

def findPeakOnsetAndOffset(peaks, phasicGSR,onset=0.01,offset=0):
    
    listOfPeaks = [] #initialize the list of Peaks
    onsetIndex = 0;
    offsetIndex = len(phasicGSR);
    for peakLocation in peaks: 
        for i in range(peakLocation, 0, -1):
            if(phasicGSR[i] <= onset):
                onsetIndex = i;
                break;
            
        for i in range(peakLocation, len(phasicGSR), 1): 
            if(phasicGSR[i] <= offset):
                offsetIndex = i;
                break;
        
        listOfPeaks.append([onsetIndex,peakLocation,offsetIndex]) #create the peak element
    return(listOfPeaks)

peaks = findPeakOnsetAndOffset(peaks4, filteredList(rawdatalist), 0.01, 0)
```

```python
#Find the amplitude of each SCR

amplitudes = []

for peak in peaks:
    #resultsDict["peak"] = {"peakStart":peak[0],"peakMax":peak[1],"peakEnd":peak[2]}
    amplitude = rawdatalist[peak[1]] - rawdatalist[peak[0]]
    amplitudes.append(amplitude)
    #print ("" + str(rawdatalist[peak[1]]) + " - " + str(rawdatalist[peak[0]]) + " = " + str(amplitude))
```

```python
#Find amplitude average

amplitudes = np.array(amplitudes)
amplitudes.mean()
```
